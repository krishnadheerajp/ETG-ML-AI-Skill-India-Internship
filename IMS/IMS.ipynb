# Fetching the items from Inventory

```python
import json
import pandas as pd
fd = open('inventory.json','r')
fetch = fd.read()
fd.close()

items = json.loads(fetch)
df = pd.DataFrame(items).T
df
```

# Adding Items to Inventory

```python
item_id = str(input('Enter the id of item: '))
item_name = str(input('Enter the name of item: '))
item_price = int(input('Enter the price of item: '))
item_qty = int(input('Enter the quantity of item: '))
item_rating = float(input('Enter the rating of item: '))
item_disc = int(input('Enter the discount for the item: '))
item_cat = str(input('Enter the category of item: '))

items[item_id] = {'name': item_name, 'price': item_price, 'quantity': item_qty, 'rating': item_rating,
                  'discount': item_disc, 'category': item_cat } 

inv = json.dumps(items)
fd = open('inventory.json','w')
fd.write(inv)
fd.close()

fd = open('inventory.json','r')
fetch = fd.read()
fd.close()

items = json.loads(fetch)
df = pd.DataFrame(items).T
df
```

# Deleting Items from Inventory

```python
del_id = str(input("Enter the id of the item to delete: "))

del items[del_id]

inv = json.dumps(items)
fd = open('inventory.json','w')
fd.write(inv)
fd.close()

fd = open('inventory.json','r')
fetch = fd.read()
fd.close()

items = json.loads(fetch)
df = pd.DataFrame(items).T
df
```

# Updating the Items from Inventory

```python
item_id = str(input('Enter the id of item: '))
item_update=str(input("Enter the name of attribute to be updated: "))
item_up_name=str(input("Enter the new value of attribute: "))

if(item_update=='rating'):
    item_up_name=float(item_up_name)
elif(item_update=='name' or item_update=='category'):
    pass
else:
    item_up_name=int(item_up_name)

items[item_id][item_update] = item_up_name

inv = json.dumps(items)
fd = open('inventory.json','w')
fd.write(inv)
fd.close()

fd = open('inventory.json','r')
fetch = fd.read()
fd.close()

items = json.loads(fetch)
df = pd.DataFrame(items).T
df
```
