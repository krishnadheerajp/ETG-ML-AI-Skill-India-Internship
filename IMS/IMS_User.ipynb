# Displaying the Items from Inventory

```python
import json
import pandas as pd
fd = open('inventory.json','r')
fetch = fd.read()
fd.close()

items = json.loads(fetch)
df = pd.DataFrame(items).T
df
```

# Purchasing the Item from Inventory 

```python
ui_id = str(input("Enter the id of item: "))
ui_qty=int(input("Enter the required quantity of item: "))

avl_qty=items[ui_id]['quantity']
fd = open('sales.json','r')
fetch = fd.read()
fd.close()
sales = json.loads(fetch)
sid=str(len(sales)+1)

if(avl_qty<ui_qty):
    print("Sorry, the maximum available items are ",str(avl_qty))
    choice=str(input("Do you still want to purchase, enter your choice(Yes/No): "))
    if(choice=='Yes'):
        print("The bill details for the purchase are: ")
        print("Item name: ",items[ui_id]['name'])
        print("Item price: ",str(items[ui_id]['price']))
        cost=items[ui_id]['price']*items[ui_id]['quantity']
        cost-=float((items[ui_id]['discount']/100)*cost)
        print("Total cost: ",str(cost))
        sales[sid]={'item_name':items[ui_id]['name'],'item_cost': cost, 'item_qty':items[ui_id]['quantity'], 'item_id':ui_id}
        items[ui_id]['quantity']=0
        inv = json.dumps(items)
        fd = open('inventory.json','w')
        fd.write(inv)
        fd.close()
    else:
        print("Sorry to see you go :(")
else:   
    print("The bill details for the purchase are: ")
    print("Item name: ",items[ui_id]['name'])
    print("Item price: ",str(items[ui_id]['price']))
    cost=items[ui_id]['price']*ui_qty
    cost-=float((items[ui_id]['discount']/100)*cost)
    print("Total cost: ",str(cost))
    sales[sid]={'item_name':items[ui_id]['name'],'item_cost': cost, 'item_qty':ui_qty, 'item_id':ui_id}
    items[ui_id]['quantity']-=ui_qty
    inv = json.dumps(items)
    fd = open('inventory.json','w')
    fd.write(inv)
    fd.close()

fd = open('inventory.json','r')
fetch = fd.read()
fd.close()

items = json.loads(fetch)
df = pd.DataFrame(items).T
df
```
